# Amortización de un automóvil 

<font color="cian">**Objetivo general**</font>


+ Buscar la mejor opción para sacar un automóvil a crédito considerando tres opciones como las principales.


<font color="cian">**Objetivos específicos**</font>

+ Analizar las tres diferentes opciones que pueden considerarse como las mejores para obtener así un automóvil nuevo.

+ Realizar comparaciones entre ellas de acuerdo a las tasas de interés que manejan, así como los pagos que se harán cada mes.

+ Concluir respecto a cada variable que se requiere al pagar para entonces sacar el auto en la opción que más convenga.





<font color="blue">**Problema a amortizar**</font>

Una persona desea comprar a crédito un automóvil Ford Figo 2018 energy TM de 5 puertas 1.5 L con precio de contado de $\$208,300$ pero requiere hacerlo a crédito buscando entre tres opciones que le parecen las mejores; estas son:
+ En la agencia
+ En banco Banamex
+ En banco Santander

El plazo que se requiere es a $3$ años contando a partir de noviembre 2018.

<font color="blue">**Modelo que representa el problema**</font>

Ecuación diferencial de capitalización continua de intereses

$$\frac{d C(t)}{dt}=r\; C(t),$$

$r=$ tasa de interés instantánea


$C(t)=$ capital inicial




```python
import pandas as pd
from datetime import date
import numpy as np
from collections import OrderedDict
from dateutil.relativedelta import *
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
from datetime import date
```

```python
def amortizacion(capital, tasa, años, pmt, enganche, fecha_inicio, pagos_año):

    # inicializa las variables
    p = 1
    ini_balance = capital
    fin_balance = capital
    
    while fin_balance > 0:
        
        # Recalcular el interés basado en el balance 
        interes = round(((tasa/pagos_año) * ini_balance), 2)
        
        # Determine el pago según si este período pagará o no el préstamo
        pmt = min(pmt, ini_balance + interes)
        capital = pmt - interes
        
        # Asegúrese de que el pago adicional se ajuste si el préstamo se está pagando
        enganche = min(enganche, ini_balance - capital)
        fin_balance = ini_balance - (capital + enganche)

        yield OrderedDict([('Mes',fecha_inicio),
                           ('Periodo', p),
                           ('Balance inicial', ini_balance),
                           ('Pago', pmt),
                           ('Capital', capital),
                           ('Interes', interes),
                           ('Enganche', enganche),
                           ('Balance final', fin_balance)])
        
        # Incrementar el contador, saldo y fecha.
        p += 1
        fecha_inicio += relativedelta(months=1)
        ini_balance = fin_balance
```

```python
def amortizacion_tabla(capital, tasa, años,
                       enganche=0, pagos_año=12, fecha_inicio=date.today()):

    # El pago se mantiene constante según los términos originales del préstamo.
    pago = -round(np.pmt(tasa/pagos_año, años*pagos_año, capital), 2)
    
    # Genera la tabla y ordena las columnas resultantes para su conveniencia.
    tabla = pd.DataFrame(amortizacion(capital, tasa, años, pago,
                                     enganche, fecha_inicio, pagos_año))
    tabla = tabla[["Periodo", "Mes", "Balance inicial", "Pago", "Interes", 
                         "Capital", "Enganche", "Balance final"]]
    
    # Convertir a un objeto de fecha
    tabla["Mes"] = pd.to_datetime(tabla["Mes"])
    
    #Tabla de estadística de resumen
    fecha_final = tabla["Mes"].iloc[-1]
    res = pd.Series([fecha_final, tabla["Periodo"].count(), tasa,
                       años, capital, pago, enganche,
                       tabla["Interes"].sum()],
                       index=["Fecha de termino", "Número de pagos", "Tasa de interes", "Años", "Capital",
                             "Pago", "Enganche", "Interes total"])
    
    return tabla, res
```

** Agencia**

# Comparación

```python
tabla1, res1 = amortizacion_tabla(168723, .128, 3, enganche=0, fecha_inicio=date(2018,11,1))
tabla2, res2 = amortizacion_tabla(208300, .11, 3, enganche=0, fecha_inicio=date(2018,11,1))
tabla3, res3 = amortizacion_tabla(208300, .1325, 3, enganche=0, fecha_inicio=date(2018,11,1))
```

```python
pd.DataFrame([res1, res2, res3])
```

```python
plt.figure(figsize=(8,6))
plt.grid()
plt.plot(xA, yA, 'k', lw=2, label='Agencia')
plt.plot(xS, yS, 'r', lw=2, label='Santander')
plt.plot(xB, yB, "b", lw=2, label="Banamex")
plt.xlabel('No. de pagos')
plt.ylabel('Intereses')
plt.legend(loc='best')
```

```python
%matplotlib inline
plt.style.use('ggplot')
```

```python
fig, ax = plt.subplots(1, 1)
tabla1.plot(x='Mes', y='Balance final',lw=4, label="Agencia", ax=ax)
tabla2.plot(x='Mes', y='Balance final',alpha=1,lw=4, color="k", label="Banamex", ax=ax)
tabla3.plot(x='Mes', y='Balance final',alpha=.65, lw=4, label="Santander", ax=ax)
plt.title("Periodos de Amortización");
```

```python
def make_plot_data(tabla, res):
    """Create a dataframe with annual interest totals, and a descriptive label"""
    y = tabla.set_index('Mes')['Interes'].resample("A").sum().reset_index()
    y["Año"] = y["Mes"].dt.year
    y.set_index('Año', inplace=True)
    y.drop('Mes', 1, inplace=True)
    label="A {} años con una tasa del {}%".format(res['Años'], res["Tasa de interes"]*100)
    return y, label
    
y1, label1 = make_plot_data(tabla1, res1)
y2, label2 = make_plot_data(tabla2, res2)
y3, label3 = make_plot_data(tabla3, res3)

y = pd.concat([y1, y2, y3], axis=1)
```

```python
figsize(7,5)
fig, ax = plt.subplots(1, 1)
y.plot(kind="bar", ax=ax)

plt.legend([label1, label2, label3], loc=1, prop={'size':10})
plt.title("Pago de intereses");
```

# Conclusión

+ AGENCIA:
Consideramos que no es la mejor opción porque tienes que dar un enganche inicial y el crédito que te ofrecen es menor que los dos bancos, pero aún cuando te prestan menos, pagas casi el mismo interés total al final del plazo dado. 
+ SANTANDER:
Creemos que Santander no es la mejor opción porque la tasa de interés que maneja es más alta que la de Banamex y por lo tanto el pago de intereses totales es superior al de Banamex y de la agencia. 
+ BANAMEX:
En conclusión, este banco es la mejor opción ya que es el que ofrece una tasa de interés menor que Santander y la agencia, por lo que recomendaríamos acudir a este banco por el prestamo solicitado.
